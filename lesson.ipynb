```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy
from matplotlib import rc
rc('animation', html='jshtml')
def plot(x_coords, y_coords):
  x_coords = numpy.array(x_coords)
  y_coords = numpy.array(y_coords)
  fig = plt.figure(figsize=(10.0, 10.0))
  ax = fig.add_subplot(1, 1, 1)
  ax.set_xlim(-max(abs(x_coords)), max(abs(x_coords)))
  ax.set_ylim(-max(abs(y_coords)), max(abs(y_coords)))
  ax.tick_params(axis='x', labelsize=8)
  ax.tick_params(axis='y', labelsize=8)

  path_plot = ax.plot(x_coords[0], y_coords[0], color='red', linewidth=5)[0]
  def update_plot(step):
      path_plot.set_data(x_coords[:step], y_coords[:step])
      return path_plot
  ani = matplotlib.animation.FuncAnimation(fig=fig, func=update_plot, frames=len(x_coords), interval=90)
  return ani


x_coords = [0, 1, 1, 1, 0, 0, -1, -2, -2, -3, -2, -2, -2, -1, -2]
y_coords = [0, 0, 1, 2, 2, 1, 1, 1, 2, 2, 2, 1, 1, 1, 0]

ani = plot(x_coords, y_coords)
ani
```
